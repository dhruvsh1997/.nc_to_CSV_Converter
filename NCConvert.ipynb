```python
import xarray as xr
import numpy as np
import pandas as pd
```

```python
nc = xr.open_dataset('20110101032000-ABOM-L3S_GHRSST-SSTskin-AVHRR_D-6d_night.nc')
#nc.to_dataframe().to_csv('20110101.csv')
```

```python
df=pd.DataFrame(nc.to_dataframe())
```

```python
row=df.shape[0]
col=df.shape[1]
```

```python
part=row/10
div=[]
part
```

```python
for i in range(1,11):
    div.append(int(i*part))
div=np.array(div)
```

```python
div
```

```python
df.iloc[0:div[0],0:col].head(3)
```

```python
df.iloc[0:div[0],0:col].to_csv("20110101032000-ABOM-0-2700000-L3S.csv")
for j in range(1,10):
    st1="20110101032000-ABOM-"+str(div[j-1])+"-"+str(div[j])+"-L3S.csv"
    df.iloc[div[j-1]:div[j],0:col].to_csv(st1)
    print(st1)
```
